# Data Processing

# Importing Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import  MinMaxScaler
from sklearn.preprocessing import  StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV


from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

# Importing the Data

```python
train_data = pd.read_csv("/content/sample_data/california_housing_train.csv")

test_data = pd.read_csv("/content/sample_data/california_housing_test.csv")
```

# splitting the data

```python
x_train = train_data.drop(columns = 'median_house_value')
x_test = test_data.drop(columns = 'median_house_value')

y_train = train_data['median_house_value']
y_test = test_data['median_house_value']
```

# Fitting the Model based on chat gpt code

```python
param_grid = {
    'n_neighbors': [3, 5, 7, 9, 11],
}

# Create the KNeighborsRegressor model
knn = KNeighborsRegressor()

# Perform Grid Search
grid_search = GridSearchCV(knn, param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(x_train, y_train)

# Get the best hyperparameters
best_params = grid_search.best_params_

# Train the best model
best_knn = KNeighborsRegressor(**best_params)
print("best_knn value is:",best_knn)
best_knn.fit(x_train, y_train)

# Make predictions on the test dataset
y_test_pred = best_knn.predict(x_test)
y_train_pred = best_knn.predict(x_train)
```

# Evaluating the model
Test Data Performance

```python
mse = mean_squared_error(y_test, y_test_pred)
mae = mean_absolute_error(y_test, y_test_pred)
r2 = r2_score(y_test, y_test_pred)

print(f"Best Hyperparameters: {best_params}")
print(f"Mean Squared Error for test data: {mse}")
print("mean absolute error for test data:", mae)
print(f"R-squared for test data: {r2}")
```

Train Data performance

```python
mse_train = mean_squared_error(y_train, y_train_pred)
mae_train = mean_absolute_error(y_train, y_train_pred)
r2_train = r2_score(y_train, y_train_pred)

print(f"Best Hyperparameters: {best_params}")
print(f"Mean Squared Error for Train data:", mse_train)
print("mean absolute error for Train data:", mae_train)
print(f"R-squared for Train data:", r2_train)
```

Scatter Plots for Actual and Predicted values of train and test datasets

```python

# Create a single figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Plot for the training set
ax1.scatter(x=y_train, y=y_train_pred)
ax1.set_xlabel("Actual", fontsize=15)
ax1.set_ylabel("Predicted", fontsize=14)
ax1.set_title('Actual vs. Predicted Values (Training Set)')

# Plot for the test set
ax2.scatter(x=y_test, y=y_test_pred)
ax2.set_xlabel("Actual", fontsize=15)
ax2.set_ylabel("Predicted", fontsize=14)
ax2.set_title('Actual vs. Predicted Values (Test Set)')

plt.show()  # Display the figure with both subplots side by side
```

# Splitting the data

```python
x_train = train_data.drop(columns = 'median_house_value')
x_test = test_data.drop(columns = 'median_house_value')

y_train = train_data['median_house_value']
y_test = test_data['median_house_value']
```

# Scalling the data

```python
scaler = MinMaxScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

First To know the optimal K-value where the error rate is low to know that value I am trying to fit the model for different k-value.

```python
rmse_val=[]
for i in range(1,22):
    knn = KNeighborsRegressor(n_neighbors = i)
    knn.fit(x_train_scaled, y_train)  #fit the model
    pred=knn.predict(x_test_scaled) #make prediction on test set
    error = np.sqrt(mean_squared_error(y_test,pred)) #calculate rmse
    rmse_val.append(error) #store rmse values
    print('RMSE value for k= ' , i , 'is:', error)
```

For the k-value= 16 I am getting the lower error rate. I think for that model Information gain is Increased and the tree is built with the more purity.

Fitting the model with k-value 16

```python
knn = KNeighborsRegressor(n_neighbors=16)
knn.fit(x_train_scaled,y_train)
```

```python
y_train_pred2 = knn.predict(x_train_scaled)
y_test_pred2 = knn.predict(x_test_scaled)
```

# Evaluating the model
Test Data

```python
mse = mean_squared_error(y_test, y_test_pred2)
mae = mean_absolute_error(y_test, y_test_pred2)
r2 = r2_score(y_test, y_test_pred2)

print(f"Mean Squared Error for test data: {mse}")
print("mean absolute error for test data:", mae)
print(f"R-squared for test data: {r2}")
```

Train Data

```python
mse_train = mean_squared_error(y_train, y_train_pred2)
mae_train = mean_absolute_error(y_train, y_train_pred2)
r2_train = r2_score(y_train, y_train_pred2)

print(f"Mean Squared Error for Train data:", mse_train)
print("mean absolute error for Train data:", mae_train)
print(f"R-squared for Train data:", r2_train)
```
